# **Movie Recommendation** **System**

```python
#importing the dependencies
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#data collection
movies=pd.read_csv('/content/movies.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
selected_features=['genres','keywords','tagline','cast','director']
```

```python
for feature in selected_features:
  movies[feature]=movies[feature].fillna('')
```

```python
#combining all the 5 features
combined_features=movies['genres']+' '+movies['keywords']+' '+movies['tagline']+' '+movies['cast']+' '+movies['director']
```

```python
print(combined_features)
combined_features.shape
```

```python
#converting the text data to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature=vectorizer.fit_transform(combined_features)
print(feature)
```

```python
similarity=cosine_similarity(feature)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
#getting the input from the user
movie_name=input('Enter your favourite movie name : ')
```

```python
#creating a list with all the movies name given in the dataset
list_of_all_titles=list(movies['title'])
print(list_of_all_titles)
```

```python
find=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find)
```

```python
match=find[0]
print(match)
```

```python
#finding the index
index_of_the_movie=movies[movies.title==match]['index'].values[0]
print(index_of_the_movie)
```

```python
#similar movies list
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
#sorting the movies
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#printing the name of similar movies based on the index
print('Suggested Movies : \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movies[movies.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```

Movie Recommendation System

```python
movie_name=input('Enter your favourite movie name : ')
list_of_all_titles=list(movies['title'])
find=difflib.get_close_matches(movie_name,list_of_all_titles)
match=find[0]
index_of_the_movie=movies[movies.title==match]['index'].values[0]
similarity_score=list(enumerate(similarity[index_of_the_movie]))
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print('Suggested Movies : \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movies[movies.index==index]['title'].values[0]
  if(i<21):
    print(i,'.',title_from_index)
    i+=1
```
